```python
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def formatter(f):
    year = int(f.split("laucnty")[1][0:2])
    year = year + (2000 if year <16 else 1900 )
    df = pd.read_excel(f,skiprows=5,convert_float=False)
    df.drop(["Unnamed: 5"], axis=1, inplace=True)
    df.columns = ["LAUS", "FIPS_state","FIPS_county", \
                  "county_state", "year", "labor_force","employed", \
                 "unemployed","unemployment_rate"]
    splitNames = df.county_state.str.split(",",expand=True)
    df.FIPS_county = df.FIPS_county.apply('{0:0>3.0f}'.format)
    df.FIPS_state = df.FIPS_state.apply('{0:0>2.0f}'.format)
    df.dropna(inplace=True)
    df.index.name = "index"
    df["county"] = splitNames.loc[:,0].astype("string")
    df["state"] = splitNames.loc[:,1].astype("string")
    df["county"] = df.county.str.split("County",expand=True).loc[:,0].astype("str")
    df.drop(["county_state"],axis=1, inplace=True)
    df.dropna(inplace=True)
    df[["year"]] = int(year)
    df["geoid"] = df.FIPS_state + df.FIPS_county
    return df
```

```python
allFiles = glob.glob("./data/county-employment/*.xlsx")
df = pd.concat([ formatter(f) for f in allFiles ])
df.sa
```

```python
df.to_csv("./data/county-employment/")
```

```python
df2 = pd.read_csv("./data/county-employment/Gaz_counties_national.txt",sep="\t")
df2.rename(columns={"GEOID":'geoid'},inplace=True)
df2.geoid = df2.geoid.apply('{0:0>5}'.format)
df2.rename(columns={df2.columns[-1]: 'lng',df2.columns[-2]: 'lat'},inplace=True)
df3 = df.merge(df2,on='geoid',how='left')
```
